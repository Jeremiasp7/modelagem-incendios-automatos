```python
import os
import pandas as pd

# Filtra as pastas que terminam com "_v1"
pastas_v1 = [p for p in os.listdir() if p.endswith("_v1") and os.path.isdir(p)]

# Listas para armazenar os dados
queimados_total = []
geracoes_total = []
q_baixa = []
q_media = []
q_alta = []

for pasta in pastas_v1:
    caminho_resultados = os.path.join(pasta, "resultados")
    
    if os.path.exists(caminho_resultados):
        arquivos = [arq for arq in os.listdir(caminho_resultados) if arq.endswith(".csv")]
        
        for arq in arquivos:
            df = pd.read_csv(os.path.join(caminho_resultados, arq))
            if 'Q' in df.columns and 'G' in df.columns:
                queimados_total.append(df["Q"].iloc[0])
                geracoes_total.append(df["G"].iloc[0])

                # Checa e coleta dados por tipo de vegetação
                if 'Q_Baixa' in df.columns:
                    q_baixa.append(df["Q_Baixa"].iloc[0])
                if 'Q_Media' in df.columns:
                    q_media.append(df["Q_Media"].iloc[0])
                if 'Q_Alta' in df.columns:
                    q_alta.append(df["Q_Alta"].iloc[0])

# Cálculos estatísticos gerais
media_q = pd.Series(queimados_total).mean()
desvio_q = pd.Series(queimados_total).std()
mediana_q = pd.Series(queimados_total).median()
min_q = min(queimados_total)
max_q = max(queimados_total)
cv_q = (desvio_q / media_q) * 100
media_g = pd.Series(geracoes_total).mean()

# Médias por tipo de vegetação
media_baixa = pd.Series(q_baixa).mean()
media_media = pd.Series(q_media).mean()
media_alta = pd.Series(q_alta).mean()

# Exibição dos resultados
print(f"Média de queimados: {media_q:.2f}")
print(f"Desvio padrão de queimados: {desvio_q:.2f}")
print(f"Mediana de queimados: {mediana_q:.2f}")
print(f"Mínimo de queimados: {min_q}")
print(f"Máximo de queimados: {max_q}")
print(f"Coeficiente de Variação (CV%) dos queimados: {cv_q:.2f}%")
print(f"Número médio de gerações: {media_g:.2f}")
print(f"\nMédia de queimados em vegetação BAIXA: {media_baixa:.2f}")
print(f"Média de queimados em vegetação MÉDIA: {media_media:.2f}")
print(f"Média de queimados em vegetação ALTA: {media_alta:.2f}")
```
